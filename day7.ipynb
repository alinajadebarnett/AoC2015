```
import re

inst_re = re.compile(r'(?P<form>.*) -> (?P<dest>[a-z]+)')
bin_form_re = re.compile(r'(?P<larg>[a-z0-9]+) (?P<op>RSHIFT|AND|OR|LSHIFT) (?P<rarg>[a-z0-9]+)')
un_form_re = re.compile(r'(?P<op>NOT) (?P<arg>[a-z0-9]+)')
literal_re = re.compile(r'(?P<arg>[a-z0-9]+)')
list_of_res = [bin_form_re, un_form_re, literal_re]

s = input.split('\n')[0]
print(s)
mo = inst_re.fullmatch(s)
mo.group('form'), mo.group('dest')
```

```
def convert_inst(inst):
  mo = inst_re.fullmatch(inst)
  if mo is None:
    raise ValueError()
  formula = mo.group('form')
  destination = mo.group('dest')

  for regex in list_of_res:
    mo2 = regex.fullmatch(formula)
    if mo2 is not None:
      if regex is bin_form_re:
        return destination, mo2.group('op'), [mo2.group('larg'), mo2.group('rarg')]
      elif regex is un_form_re:
        return destination, 'NOT', [mo2.group('arg')]
      elif regex is literal_re:
        return destination, 'LITERAL', [mo2.group('arg')]
  raise ValueError()
```

```
for line in input.split('\n'):
  print(line)
  print(convert_inst(line))
```

```
import networkx as nx
```

```
def init_structures(instructions):
  recipe_dict = {}
  dependancy_graph = nx.DiGraph()
  for instruction in instructions:
    dest, op, dependancies = convert_inst(instruction)
    recipe_dict[dest] = (op, dependancies)
    dependancy_graph.add_edges_from([(dep, dest) for dep in dependancies])
  return recipe_dict, dependancy_graph

value_store = {}
recipe_dict, dependancy_graph = init_structures(input.split('\n'))
```

```
import numpy as np

def run_sim(instructions):
  value_store = {}
  recipe_dict, dependancy_graph = init_structures(instructions.split('\n'))
  for w in nx.topological_sort(dependancy_graph):
    recipe_w = recipe_dict.get(w, None)
    if recipe_w is None:
      value_store[w] = np.uint16(w)
    else:
      ops, dependancies = recipe_w
      if ops == 'RSHIFT':
        l = value_store[dependancies[0]]
        r = value_store[dependancies[1]]
        value_store[w] = l >> r
      elif ops == 'AND':
        l = value_store[dependancies[0]]
        r = value_store[dependancies[1]]
        value_store[w] = l & r
      elif ops == 'OR':
        l = value_store[dependancies[0]]
        r = value_store[dependancies[1]]
        value_store[w] = l | r
      elif ops == 'LSHIFT':
        l = value_store[dependancies[0]]
        r = value_store[dependancies[1]]
        value_store[w] = l << r
      elif ops == 'NOT':
        r = value_store[dependancies[0]]
        value_store[w] = ~r
      elif ops == 'LITERAL':
        value_store[w] = value_store[dependancies[0]]
      else:
        print('got here', ops)
  return value_store
```

```
test_input = '''123 -> x
456 -> y
x AND y -> d
x OR y -> e
x LSHIFT 2 -> f
y RSHIFT 2 -> g
NOT x -> h
NOT y -> i'''
run_sim(test_input)
```
