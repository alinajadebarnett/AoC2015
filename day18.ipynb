```
import numpy as np
from scipy.signal import convolve2d
```

```
def read_symbols(input):
  return np.array([[x == '#' for x in line] for line in input.split('\n')], dtype=np.uint8)

init_grid = read_symbols(input)
init_grid[-9:, -9:]
```

```
kernel = np.ones((3,3,), dtype=np.uint8)
kernel[1,1] = 0
kernel
```

```
def count_neighbors(grid):
  return convolve2d(grid, kernel, mode='same')

print(test_inputs[0])
print()
print(count_neighbors(read_symbols(test_inputs[0])))
```

```
def step(grid):
  nn = count_neighbors(grid)
  nne3 = nn == 3
  nne2o3 = nne3 | (nn == 2)
  ison = grid == 1
  ret_as_bool = (ison & nne2o3) | (~ison & nne3)
  return ret_as_bool.astype(np.uint8)

np.array_equal(step(step(read_symbols(test_inputs[0]))), read_symbols(test_inputs[2]))
```

```
grid = init_grid
for _ in range(100):
  grid = step(grid)
grid.sum()
```

```
def set_corners_on(grid):
  ret = grid.copy()
  for i in (0, -1):
    for j in (0, -1):
      ret[i, j] = 1
  return ret
```

```
def step2(grid):
  nn = count_neighbors(grid)
  nne3 = nn == 3
  nne2o3 = nne3 | (nn == 2)
  ison = grid == 1
  ret_as_bool = (ison & nne2o3) | (~ison & nne3)
  return set_corners_on(ret_as_bool.astype(np.uint8))
```

```
grid = set_corners_on(init_grid)
for _ in range(100):
  grid = step2(grid)
grid.sum()
```
