```
def hash_good_enough(s, n):
  s += str(n)
  s = bytes(s, 'ascii')
  first5 = md5(s).hexdigest()[:5]
  return first5 == '00000'

print(hash_good_enough('abcdef', 609043))
print(hash_good_enough('abcdef', 609042))
```

```
from itertools import count

def mine_coins(s):
  for i in count(1):
    if hash_good_enough(s, i):
      return i

print(mine_coins('abcdef'))
print(mine_coins('pqrstuv'))
print(mine_coins('bgvyzdsv'))
```

```
def hash_gooder_enough(s, n):
  s += str(n)
  s = bytes(s, 'ascii')
  first6 = md5(s).hexdigest()[:6]
  return first6 == '000000'
```

```
def mine_better_coins(s):
  for i in count(1):
    if hash_gooder_enough(s, i):
      return i


print(mine_better_coins('bgvyzdsv'))
```
